```python
# Complexity O(n^2)

from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        result = []
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    result.append(i)
                    result.append(j)
                    break
        return result
```

```python
# Instantiate the Solution class
solution = Solution()

# Use the twoSum function
nums = [2, 7, 11, 15]
target = 9
result = solution.twoSum(nums, target)

print(f"Indexes: {result}")
print(f"Numbers: {nums[result[0]]} + {nums[result[1]]} = {target}")
```

```python
# Complexity O(n) (using a hash map, solution optimized with dict from leetcode, solution made by ChatGPT)

from typing import List

class Solution:
    def two_sum_optimized(self, nums: List[int], target: int) -> List[int]:
        
        # Dict to save seen numbers and their indices: {value: index}
        seen = {}

        for i, num in enumerate(nums):
            
            complement = target - num
            
            # Check if the complement number needed to reach the target has already been seen
            if complement in seen:
                # If yes, return the index of the complement and the current index
                return [seen[complement], i]
            
            # If not, save the current number and its index in the dictionary
            seen[num] = i
        
        return [] # Return empty if no solution is found
```

```python
#nums = [2, 7, 11, 15]
#target = 9

nums = [2, 11, 15, 7]
target = 9

solution = Solution()
result = solution.two_sum_optimized(nums, target)

print(f"Indexes: {result}")
print(f"Numbers: {nums[result[0]]} + {nums[result[1]]} = {target}")

#nums = [3,2,4]
#target = 6

#nums = [3,3]
#target = 6
```

```python
def verify_two_sum_solution():
    test_cases = [
        ([2, 7, 11, 15], 9, [0, 1]),
        ([2, 11, 15, 7], 9, [0, 3]),
        ([3, 2, 4], 6, [1, 2]),
        ([3, 3], 6, [0, 1]),
    ]
    
    solution = Solution()
    
    for nums, target, expected in test_cases:
        result = solution.two_sum_optimized(nums, target)
        assert result == expected, f"Failed for nums: {nums}, target: {target}. Expected: {expected}, Got: {result}"
    
    print("All test cases passed!")
```
